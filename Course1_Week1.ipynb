#Hello World of Tensorflow Sample Code
Here you will notice that we did the following things to train and predict arbritary model:


1.   Import all of the libraries
2.   Define the neural network
3.   Compile the neural network
4.   Provide the data
5.   Train the model
6.   Predict for a certain x value



```
#Import all of the libraries
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

```
#Define the neural network
model = tf.keras.Sequential([keras.layers.Dense(units=1,input_shape=[1])])
```

```
#Compile the neural network
model.compile(optimizer='sgd',loss='mean_squared_error')
```

```
#Provide the data
xs = np.array([-1.0,  0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
```

```
#Train the models
model.fit(xs, ys, epochs=500)
```

```
#Predict for 10
print(model.predict([10.0]))
```

So, for this exercise you will write code that does a similar task -- in this case predicting house prices based on a simple, linear equation.

```
#import all of the libraries
import tensorflow as tf
import numpy as np 
from tensorflow import keras
```

```
#Define the neural network
montreal = tf.keras.Sequential([keras.layers.Dense(units=1,input_shape=[1])])
```

```
#Compile the neural network
montreal.compile(optimizer='sgd', loss='mean_squared_error')
```

```
#Provide the data
xm = np.array([1.0, 2.0, 3.0, 4.0], dtype=float)
ym = np.array([430000.0, 435000.0, 440000.0, 445000.0], dtype=float)
```

```
#Train the model
montreal.fit(xm, ym, epochs=500)
```

```
#Predict when 5th year
print(montreal.predict([5.0]))
```
